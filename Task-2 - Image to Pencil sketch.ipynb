# Task-2: Image to Pencil Sketch

### In this task as per the requirement I just converted the colored image to the pencil sketch..

Following are the steps required to convert the image to the pencil sketch, In this task I have used a picture of the CAT to demonstrate-

* Importing the neccessary libraries like OpenCV and Matplotlib
* Converting the image to the Gray scale image
* Inverting the Gray scale image'
* Blurring the image using gaussian function of Opencv
* Inverting the blurred image and showing the Pencil sketched image
* Comparing the original image and pencil sketched image

### Importing the neccessary libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Reading the Original Image

```python
img = cv2.imread("cat_50_1_50.jpg")
cv2.imshow("Cat", img)
cv2.waitKey(0)
```

```python
plt.imshow(img) #Displaying the image
```

```python
color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(color_img)
```

### Converting the image to gray scale

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Cat in black and white", gray_img)
cv2.waitKey(0)
```

```python
plt.imshow(gray_img)
```

### Inverting the image

```python
inverted_img = 255 - gray_img
cv2.imshow("Inverted", inverted_img)
cv2.waitKey()
```

```python
plt.imshow(inverted_img)
```

### Now lets Blurr the image using gaussian function 

```python
blurred_img = cv2.GaussianBlur(inverted_img, (21, 21), 0)
plt.imshow(blurred_img)
```

### Final Step - Inverting the blurred image

```python
inverted_blurred_img = 255 - blurred_img
pencil_sketch = cv2.divide(gray_img, inverted_blurred_img, scale=256.0)
cv2.imshow("Pencil Sketch", pencil_sketch)
cv2.waitKey(0)
```

### Displaying the Pencil sketched image

```python
plt.imshow(pencil_sketch)
```

```python
cv2.imshow("original image", img)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```

## Conclusion

The image has been successfully converted into Pencil Sketched Image
